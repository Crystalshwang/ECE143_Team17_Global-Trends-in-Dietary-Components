```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
## mean of daily caloric supply
df=pd.read_csv('daily-caloric-supply-derived-from-carbohydrates-protein-and-fat.csv')#,encoding = "ISO-8859-1")
df.head()
name = ['Animal protein','Plant protein','Fat','Carbohydrates']   #labels
theta = np.linspace(0,2*np.pi,len(name),endpoint=False)    #divide the circle to uniform part
value = [df[i].mean() for i in df.columns[3:]] # mean kilocalories in different macronutrient
print(value)
theta = np.concatenate((theta,[theta[0]]))  #close
value = np.concatenate((value,[value[0]]))  #close the circle

ax = plt.subplot(111,projection = 'polar')      
ax.plot(theta,value,'r-',lw=1,alpha = 0.75)    
ax.fill(theta,value,'r',alpha = .55)           #fill, alpha is transparency
ax.set_thetagrids(theta*180/np.pi,name,fontsize=12)         #label
ax.set_ylim(0,2000)                          #set polar axis length
ax.set_theta_zero_location('S')         #set polar axis direction
ax.set_title('daily caloric supply',fontsize = 15,pad=20.0)   #description of plot
plt.show()
```

```python
df=pd.read_csv('daily-caloric-supply-derived-from-carbohydrates-protein-and-fat.csv')#,encoding = "ISO-8859-1")
df.head()
```

```python
grp = df.groupby('Entity')
res=grp.mean()
res
```

```python
countries = ['Australia','Canada','Japan', 'Brazil', 'China','United Kingdom', 'France', 'Egypt', 'Mexico','United States']
new_df=res.loc[countries,:]
new_df.drop('Year',axis=1,inplace=True)
new_df
```

```python
ax=new_df.plot.barh(stacked=True, alpha=0.5)
ax.legend(['Animal protein','Plant protein','Fat','Carbohydrates'],loc='best')
ax.set(xlim=[0, 5000])
plt.xlabel('kilocalories per person per day')
# choose 6 countries to represent different continents and map their diet compositions
```
